Chatbot.jsx

```python
import React, { useState, useEffect, useRef } from 'react';
import './Chatbot.css'; // <--- Importamos los estilos aqu√≠

const Chatbot = () => {
  // --- ESTADO (Hooks) ---
  const [isOpen, setIsOpen] = useState(false);
  const [inputValue, setInputValue] = useState("");
  const [messages, setMessages] = useState([]);
  const [isProcessing, setIsProcessing] = useState(false);

  // Referencia para scroll autom√°tico
  const messagesEndRef = useRef(null);

  const scrollToBottom = () => {
    messagesEndRef.current?.scrollIntoView({ behavior: "smooth" });
  };

  useEffect(() => {
    scrollToBottom();
  }, [messages, isProcessing]);

  // --- L√ìGICA ---
  const handleSendMessage = (text) => {
    if (!text.trim()) return;

    // 1. Agregar mensaje del usuario
    const newUserMessage = { text: text, sender: 'user' };
    setMessages((prev) => [...prev, newUserMessage]);
    setInputValue("");
    setIsProcessing(true);

    // 2. Simular llamada a API
    setTimeout(() => {
      const botResponse = { 
        text: "Esta es una respuesta simulada desde la API.", 
        sender: 'bot' 
      };
      setMessages((prev) => [...prev, botResponse]);
      setIsProcessing(false);
    }, 2000);
  };

  const suggestedQuestions = [
    "Please do my staff planning...",
    "Bring me the anomalies...",
    "Compare monthly On Time openings...",
    "Give me the Mystery Shopper Score..."
  ];

  // --- RENDERIZADO ---
  return (
    <div className="chatbot-container">
      
      {/* Bot√≥n flotante (Launcher) */}
      {!isOpen && (
        <button 
          onClick={() => setIsOpen(true)} 
          className="chatbot-launcher"
        >
          üí¨
        </button>
      )}

      {/* Ventana de Chat */}
      {isOpen && (
        <div className="chat-window">
          
          {/* Header */}
          <div className="chat-header">
            <div className="header-title">
              <span>ü§ñ</span> 
              <strong>Edgar</strong>
            </div>
            <button onClick={() => setIsOpen(false)} className="close-btn">
              ‚úñ
            </button>
          </div>

          {/* Cuerpo */}
          <div className="chat-body">
            
            {/* Bienvenida */}
            {messages.length === 0 && (
              <div className="welcome-message">
                <p>Hello, ask me anything!</p>
              </div>
            )}

            {/* Sugerencias */}
            {messages.length === 0 && (
              <div className="suggestions-container">
                {suggestedQuestions.map((q, index) => (
                  <button 
                    key={index} 
                    className="suggestion-btn"
                    onClick={() => handleSendMessage(q)}
                  >
                    {q}
                  </button>
                ))}
              </div>
            )}

            {/* Historial de Mensajes */}
            {messages.map((msg, index) => (
              <div 
                key={index} 
                className={`message-bubble ${msg.sender}`}
              >
                {msg.text}
              </div>
            ))}

            {/* Mensaje "Procesando..." */}
            {isProcessing && (
              <div className="message-bubble processing">
                Procesando...
              </div>
            )}
            <div ref={messagesEndRef} />
          </div>

          {/* Footer (Input) */}
          <div className="chat-footer">
            <input
              type="text"
              placeholder="Type your message..."
              value={inputValue}
              onChange={(e) => setInputValue(e.target.value)}
              onKeyPress={(e) => e.key === 'Enter' && handleSendMessage(inputValue)}
              className="chat-input"
              disabled={isProcessing}
            />
            <button 
              onClick={() => handleSendMessage(inputValue)} 
              className="send-btn"
              disabled={isProcessing}
            >
              ‚û°
            </button>
          </div>
          <div className="chat-disclaimer">
            AI-generated content may be inaccurate. Terms and Conditions.
          </div>
        </div>
      )}
    </div>
  );
};

export default Chatbot;
```
